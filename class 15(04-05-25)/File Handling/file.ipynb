* what is OS module

```python
import os

def check_file_or_folder(name):
    if os.path.isfile(name):
        print("it is a file")
    elif os.path.isdir(name):
        print("it is a folder")
    else:
        print("it is not a file or folder")

name=input("check the is file or a folder:")
check_file_or_folder(name)
```

# Learn About How PIL and numpy module work

```python
from PIL import Image

img = Image.open("datatypes1.png")

img.show()


# Image ke bare me jankari
print(f"Size: {img.size}")  # (width, height)
print(f"Format: {img.format}")  # JPEG, PNG etc.
print(f"Mode: {img.mode}")  # RGB, L (grayscale), etc.
```

*  BY PIL we crop any image and save again

```python
from PIL import Image

img = Image.open("datatypes1.png")

img.show()
# New size (width, height)
new_size = (300, 200)

# Resize karein
resized_img = img.resize(new_size)

# Save karein
resized_img.save("resized_image.png")
```

```python
# we can rotate the image by 90 degrees
img = Image.open("datatypes1.png")
# 45 degree rotate karein
rotated_img = img.rotate(90)
rotated_img.show()
img.close()
```

# Image me Text Add Karna

```python
from PIL import Image, ImageDraw, ImageFont

img = Image.open("datatypes1.png")

# Draw object banayein
draw = ImageDraw.Draw(img)

# Font select karein (size 30)
font = ImageFont.truetype("arial.ttf", 60)

# Text add karein (position, text, color, font)
draw.text((50, 50), "Hello World!", (255, 0, 0), font=font)

# show image karein
img.show()

# Save karein
img.save("text_image.png")
```

# 7. Image Filter Lagana

```python
from PIL import Image, ImageFilter

img = Image.open("datatypes1.png")

# Blur filter lagayein
blurred_img = img.filter(ImageFilter.BLUR)

# Save karein
blurred_img.save("blurred_image.png")

# Edge enhance filter
edge_img = img.filter(ImageFilter.EDGE_ENHANCE)
edge_img.save("edge_image.png")
```

# 8. Image Convert Karna (JPG to PNG)

```python
from PIL import Image

img = Image.open("datatypes1.png")

# PNG me convert karke save karein
img.save("converted_image.jpg", "JPEG")
```

# 9. Multiple Images ka Collage Banana

```python
from PIL import Image

# New blank image banayein (white background)
collage = Image.new('RGB', (800, 600), (255, 255, 255))

# Images open karein
img1 = Image.open("datatypes1.png").resize((300, 200))
img2 = Image.open("datatypes1.png").resize((300, 200))

# Images ko collage me paste karein
collage.paste(img1, (50, 50))
collage.paste(img2, (400, 50))

#show karein
collage.show()
# Save karein
collage.save("datatypes1.png")
```
