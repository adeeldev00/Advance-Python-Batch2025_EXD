###  write a decorator measure_time that prints how long a function took to run.test it on a function simulate() that waits for 2 seconds and prints "simulation done!

```python
# write a decorator measure_time that prints how long a function took to run.test it on a function simulate() that waits for 2 seconds and prints "simulation done!


# Time deccorator

import time

def measure_time(func):
    def wrapper():
        print("Function has start here!")
        start_time = time.time()
        
        result = func()
        
        end_time = time.time()
        print("Function end here!")
        print(f"Execution Time: {end_time - start_time:.6f} seconds")
        
        return result
    return wrapper

@measure_time
def simulate():
    time.sleep(2)
    print("Simulation done!")
simulate()
```

```python
'''
we create a decporator that add a  fancy border around the output of a function.

'''
def ascii_art(func):
    def wrapper(*args, **kwargs):
        result = func(*args, **kwargs)  
        border_length = len(result)  
        print("*" * border_length)
        print(result)  
        print("*" * border_length)
         
    return wrapper

@ascii_art
def greet(Fullname="Adeel"):
    return f"Hello, {Fullname}!"  # return a string

greet()
greet("Ali")
greet("Sara")
greet("John Doe")
greet("Jane Smith")
greet("Alice Johnson")


```

```python
# Static Vs Class Method

class Geometry:
    dis = "i am Geometry class"
    @staticmethod
    def area_of_circle(radius):
        return 3.14 * radius * radius

    @classmethod
    def description(cls):
        print(f"This is a class method of {cls.__name__} class.")
        return cls.dis

print(Geometry.area_of_circle(5))  # Accessing static method
print(Geometry.description())
```

```python
# To Find Which Month Has Highest Flights

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load flights dataset
flights = sns.load_dataset("flights")

# Group by 'month' and calculate total passengers
monthly_total = flights.groupby("month")["passengers"].sum().reset_index()

# Optional: Sort months in correct calendar order
month_order = ["January", "February", "March", "April", "May", "June", 
               "July", "August", "September", "October", "November", "December"]
monthly_total["month"] = pd.Categorical(monthly_total["month"], categories=month_order, ordered=True)
monthly_total = monthly_total.sort_values("month")

# Plot line chart
plt.figure(figsize=(10, 6))
sns.lineplot(data=monthly_total, x="month", y="passengers", marker="o")

plt.title("Total Passengers per Month (Across All Years)")
plt.xlabel("Month")
plt.ylabel("Total Passengers")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load tips dataset
tips = sns.load_dataset("tips")

# Group by gender and calculate total tips
gender_tips = tips.groupby("sex")["tip"].sum().reset_index()

# Optional: Sort for neatness
gender_tips = gender_tips.sort_values("tip")

# Plotting line chart
plt.figure(figsize=(8, 5))
sns.lineplot(data=gender_tips, x="sex", y="tip", marker="o", linewidth=2)

plt.title("Total Tips Given by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Tips")
plt.grid(True)
plt.tight_layout()
plt.show()
```
